```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('pizza_sales_data.csv', parse_dates=['date'])

# Add month column
df['month'] = df['date'].dt.month

# Total revenue
print("Total Revenue:", df['total_price'].sum())

# Pizzas per order (average)
print("Avg Pizzas per Order:", df['quantity'].mean())

# Daily revenue trend
daily_revenue = df.groupby('date')['total_price'].sum()
daily_revenue.plot(figsize=(12, 4), title="Daily Revenue Trend")

# Top selling pizzas
top_pizzas = df.groupby('pizza_type')['quantity'].sum().sort_values(ascending=False)
top_pizzas.plot(kind='bar', title="Top Selling Pizzas", ylabel="Total Sold")


```
